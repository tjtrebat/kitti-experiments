```python
base = Path.home() / "kitti"
train_dir = base / "training"
train_labels_dir = train_dir / "label_2"
kitti_train_labels = sorted(train_labels_dir.glob("*.txt"))
```

```python
object_types = []
extents = []
for label_file in kitti_train_labels:
    labels = utils.parse_label_file(label_file)
    for label in labels:
        object_types.append(label["type"])
        extents.append(label["dimensions"])
```

```python
df = pd.DataFrame({
    'type': object_types,
    'height': [extent[0] for extent in extents],
    'width': [extent[1] for extent in extents],
    'length': [extent[2] for extent in extents],
})
stats = df.groupby('type').agg(['mean', 'std'])
stats
```
