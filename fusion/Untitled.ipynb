```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
configs = ['Axis-Aligned', 'Oriented']
metrics = ['Prec.', 'Rec.', 'F1']

# 3D IoU and BEV IoU values
iou3d = np.array([
    [0.227, 0.270, 0.246],
    [0.320, 0.201, 0.247]
])
beviou = np.array([
    [0.303, 0.360, 0.329],
    [0.465, 0.292, 0.359]
])

x = np.arange(len(metrics))  # the label locations
width = 0.35  # width of the bars

# Create vertical subplots
fig, axes = plt.subplots(2, 1, figsize=(7, 6), sharex=True)

# --- 3D IoU plot ---
axes[0].bar(x - width/2, iou3d[0], width, label='Axis-Aligned', color='skyblue')
axes[0].bar(x + width/2, iou3d[1], width, label='Oriented', color='lightgreen')
axes[0].set_title('3D IoU')
axes[0].legend()
axes[0].grid(axis='y', linestyle='--', alpha=0.6)

# --- BEV IoU plot ---
axes[1].bar(x - width/2, beviou[0], width, label='Axis-Aligned', color='skyblue')
axes[1].bar(x + width/2, beviou[1], width, label='Oriented', color='lightgreen')
axes[1].set_title('BEV IoU')
axes[1].legend()
axes[1].grid(axis='y', linestyle='--', alpha=0.6)

# Shared x-axis labels
axes[1].set_xticks(x)
axes[1].set_xticklabels(metrics)

# --- Figure styling ---
fig.suptitle('Comparison of BEV and 3D IoU (Ground-Truth 2D Detections, 2000 Samples)', fontsize=12)
fig.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
